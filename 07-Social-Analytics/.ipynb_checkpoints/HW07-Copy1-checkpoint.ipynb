```python
# Target User Accounts
compound_list = []
positive_list = []
negative_list = []
neutral_list = []
tweet_idx = []
media_list=[]
media_names=[]
date_list=[]
target_user = ("BBC","CBS", "CNN",  "fox","nytimes")
now=datetime.now()
#target_user = ("nytimes",)

#target_user = ("BBC","CBS")
# Loop through each user
for user in target_user:
    # Variables for holding sentiments
    ctr=0

    # Loop through 10 pages of tweets (total 200 tweets)
    for x in range(1):

        # Get all tweets from home feed
        #public_tweets = api.user_timeline(user,count = 100)
        public_tweets = api.user_timeline(user, count = 100, include_rts = True)

        # Loop through all tweets
        for tweet in public_tweets:
            #print (tweet)
            #time.strptime(tweet['created_at'],'%a %b %d %H:%M:%S +0000 %Y')
            then=datetime.strptime(tweet['created_at'], '%a %b %d %H:%M:%S +0000 %Y')
            #print(f"{tweet['user']['screen_name']} {now} / {then} / {(now-then).total_seconds()}")
         
            # Run Vader Analysis on each tweet
            compound = analyzer.polarity_scores(tweet["text"])["compound"]
            pos = analyzer.polarity_scores(tweet["text"])["pos"]
            neu = analyzer.polarity_scores(tweet["text"])["neu"]
            neg = analyzer.polarity_scores(tweet["text"])["neg"]

            # Add each value to the appropriate array
            compound_list.append(compound)
            positive_list.append(pos)
            negative_list.append(neg)
            neutral_list.append(neu)
            media_list.append(tweet["user"]['screen_name'])
            media_names.append(tweet["user"]['name'])
            date_list.append(then)
            #timestamp_list.append(tweet['created_at'])
            tweet_idx.append(ctr)
            ctr+=1
```

```python
my_colors = ["lightblue","green","red","blue","yellow"]
user_df=pd.DataFrame({"Media Source":media_list,
                          "Compound":compound_list,
                             "Positive":positive_list,
                             "Negative":negative_list,
                             "Neutral":neutral_list,
                             "Tweets Ago":tweet_idx,
                             "Name":media_names,
                             "Created":date_list
                             }
                            )
```
